```python
import matplotlib.pyplot as plt
import pandas as pd

# Create sample data
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
       'Age': [25, 32, 41, 28],
       'Score': [85, 92, 78, 88]}

df = pd.DataFrame(data)

# Create bar chart
plt.figure(figsize=(8, 6))
df.plot(kind='bar', x='Name', y='Score', rot=0)
plt.title('Scores by Name')
plt.xlabel('Name')
plt.ylabel('Score')
plt.show()
```
